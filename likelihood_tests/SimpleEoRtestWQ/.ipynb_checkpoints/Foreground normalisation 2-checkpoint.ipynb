Import libs

```python
import numpy
import numpy as np
from numpy import * 
import pylab
import pylab as P
import scipy
from scipy import optimize
from scipy.optimize import curve_fit
from scipy import misc
from scipy import ndimage
```

1/3 deg. resolution sim. normalised in the image-domain with respect to the 1/3 deg. resolution GSM median std between -25 and -35 deg declination.

Define constants: GSM normalisation, FoV, resolution etc.

```python

Tb_experimental_std_K = 62.0
FoV_deg = 12.34 #12.0 degrees is the required FoV for the foregrounds (matching the 12.0407656283 deg. of the 128*128 pix EoR sim)
lres_deg = 1./3 #1/3 deg Haslam map
lres_rad = lres_deg*(np.pi/180.)
N_lr = int(FoV_deg/lres_deg) #FoV_deg/start_res = 36
print N_lr
d_Omega_lr = lres_rad**2.
duv_lr = 1./d_Omega_lr

N_hr = 127
hres_deg = FoV_deg/N_hr
hres_rad = hres_deg*(np.pi/180.)
print N_hr
print hres_deg #EoR sim. is 0.09406848147109376 deg.
d_Omega_hr = hres_rad**2.
duv_hr = 1./d_Omega_hr
```

Generate a GRN realisation map and fft to the uv-domain

```python
wn_im_lr=np.random.normal(0,10,N_lr**2).reshape(N_lr,N_lr)
axes_tuple=(0,1)
fft_wn_im_lr=numpy.fft.ifftshift(wn_im_lr, axes=axes_tuple)
fft_wn_im_lr=numpy.fft.fftn(fft_wn_im_lr, axes=axes_tuple) #FFT (python pre-normalises correctly! -- see parsevals theorem for discrete fourier transform.)
fft_wn_im_lr=numpy.fft.fftshift(fft_wn_im_lr, axes=axes_tuple)
fft_wn_im_lr = d_Omega_lr*fft_wn_im_lr #Apply physical normalisation for FFT as per Beardsley doc.

P.imshow(log10(abs(fft_wn_im_lr)))
P.colorbar()
P.show()
```

Zero-pad the uv-plane and ifft back to the image

```python
zero_padded_fft_wn_im_lr = np.zeros([N_hr,N_hr])+0.0j
padding_width = (N_hr-fft_wn_im_lr.shape[0])/2
print padding_width, N_hr, fft_wn_im_lr.shape[0]
zero_padded_fft_wn_im_lr[padding_width:-padding_width,padding_width:-padding_width] = fft_wn_im_lr

P.imshow(np.log10(abs(zero_padded_fft_wn_im_lr)))
P.colorbar()
P.show()

wn_im_hr_test=np.random.normal(0,10,N_hr**2).reshape(N_hr,N_hr)
axes_tuple=(0,1)
fft_wn_im_hr=numpy.fft.ifftshift(wn_im_hr_test, axes=axes_tuple)
fft_wn_im_hr=numpy.fft.fftn(fft_wn_im_hr, axes=axes_tuple) #FFT (python pre-normalises correctly! -- see parsevals theorem for discrete fourier transform.)
fft_wn_im_hr=numpy.fft.fftshift(fft_wn_im_hr, axes=axes_tuple)
fft_wn_im_hr = d_Omega_hr*fft_wn_im_hr #Apply physical normalisation for FFT as per Beardsley doc.

print fft_wn_im_hr[N_hr/2,N_hr/2]
print zero_padded_fft_wn_im_lr[N_hr/2,N_hr/2]

```

```python
zero_padded_wn_im_lr=numpy.fft.ifftshift(zero_padded_fft_wn_im_lr, axes=axes_tuple)
zero_padded_wn_im_lr=numpy.fft.ifftn(zero_padded_wn_im_lr, axes=axes_tuple) #FFT (python pre-normalises correctly! -- see parsevals theorem for discrete fourier transform.)
zero_padded_wn_im_lr=numpy.fft.fftshift(zero_padded_wn_im_lr, axes=axes_tuple)
zero_padded_wn_im_lr=duv_hr*zero_padded_wn_im_lr

print duv_hr/duv_lr
print 'wn_im_lr.shape, wn_im_lr.std():', wn_im_lr.shape, wn_im_lr.std()
print 'zero_padded_wn_im_lr.shape, zero_padded_wn_im_lr.std():',zero_padded_wn_im_lr.shape, zero_padded_wn_im_lr.std()
print 'zero_padded_wn_im_lr.shape, zero_padded_wn_im_lr.real.std():',zero_padded_wn_im_lr.shape, zero_padded_wn_im_lr.real.std()

P.figure()
P.imshow(wn_im_lr)
P.colorbar()

P.figure()
P.imshow(zero_padded_wn_im_lr.real)
P.colorbar()
P.show()
#print zero_padded_wn_im_lr.real
print zero_padded_wn_im_lr.imag
```

Normalise both the WN image and the zero-padded WN image using the GSM std.

```python
gsm_normalisation_factor = (Tb_experimental_std_K/wn_im_lr.std())
normalised_wn_im_lr = wn_im_lr* gsm_normalisation_factor
normalised_zero_padded_wn_im_lr = zero_padded_wn_im_lr* gsm_normalisation_factor

P.figure()
P.imshow(normalised_wn_im_lr)
P.colorbar()

P.figure()
P.imshow(normalised_zero_padded_wn_im_lr.real)
P.colorbar()
P.show()
print duv_hr/duv_lr
print 'normalised_wn_im_lr.shape, normalised_wn_im_lr.std():', normalised_wn_im_lr.shape, normalised_wn_im_lr.std()
print 'normalised_zero_padded_wn_im_lr.shape, normalised_zero_padded_wn_im_lr.std():',normalised_zero_padded_wn_im_lr.shape, normalised_zero_padded_wn_im_lr.std()
```

Transform normalised image and zero padded image back to the uv to compare and confirm agreement

```python

normalised_zero_padded_fft_wn_im_lr=numpy.fft.ifftshift(normalised_zero_padded_wn_im_lr, axes=axes_tuple)
normalised_zero_padded_fft_wn_im_lr=numpy.fft.fftn(normalised_zero_padded_fft_wn_im_lr, axes=axes_tuple) #FFT (python pre-normalises correctly! -- see parsevals theorem for discrete fourier transform.)
normalised_zero_padded_fft_wn_im_lr=numpy.fft.fftshift(normalised_zero_padded_fft_wn_im_lr, axes=axes_tuple)
normalised_zero_padded_fft_wn_im_lr=d_Omega_hr*normalised_zero_padded_fft_wn_im_lr

normalised_fft_wn_im_lr=numpy.fft.ifftshift(normalised_wn_im_lr, axes=axes_tuple)
normalised_fft_wn_im_lr=numpy.fft.fftn(normalised_fft_wn_im_lr, axes=axes_tuple) #FFT (python pre-normalises correctly! -- see parsevals theorem for discrete fourier transform.)
normalised_fft_wn_im_lr=numpy.fft.fftshift(normalised_fft_wn_im_lr, axes=axes_tuple)
normalised_fft_wn_im_lr=d_Omega_lr*normalised_fft_wn_im_lr

P.figure()
P.imshow(np.log10(abs(normalised_fft_wn_im_lr)))
P.colorbar()

P.figure()
P.imshow(np.log10(abs(normalised_zero_padded_fft_wn_im_lr))[N_hr/2-N_lr/2:N_hr/2+N_lr/2+1,N_hr/2-N_lr/2:N_hr/2+N_lr/2+1])
P.colorbar()

P.figure()
P.imshow(np.log10(abs(normalised_zero_padded_fft_wn_im_lr))[N_hr/2-N_lr/2:N_hr/2+N_lr/2+1,N_hr/2-N_lr/2:N_hr/2+N_lr/2+1] - np.log10(abs(normalised_fft_wn_im_lr)))
P.colorbar()
P.show()

wn_im_hr_test=np.random.normal(0,10,N_hr**2).reshape(N_hr,N_hr)
axes_tuple=(0,1)
fft_wn_im_hr=numpy.fft.ifftshift(wn_im_hr_test, axes=axes_tuple)
fft_wn_im_hr=numpy.fft.fftn(fft_wn_im_hr, axes=axes_tuple) #FFT (python pre-normalises correctly! -- see parsevals theorem for discrete fourier transform.)
fft_wn_im_hr=numpy.fft.fftshift(fft_wn_im_hr, axes=axes_tuple)
fft_wn_im_hr = d_Omega_hr*fft_wn_im_hr #Apply physical normalisation for FFT as per Beardsley doc.
# print (np.log10(abs(normalised_zero_padded_fft_wn_im_lr))[N_hr/2-N_lr/2:N_hr/2+N_lr/2+1,N_hr/2-N_lr/2:N_hr/2+N_lr/2+1] - np.log10(abs(normalised_fft_wn_im_lr)))
```
